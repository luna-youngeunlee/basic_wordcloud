## 5) 워드 클라우드

① 인터넷에서 좋은 글(영문으로 된 연설문, 혹은 기사, 책 내용)을 찾아, txt로 저장하고 워드 클라우드를 만들어 봅시다.

* 코드는 예제파일을 참조하세요.
* 불필요한 단어/조사는 제거하세요.
* 글의 내용을 한눈에 파악할 수 있는 멋진 워드클라우드를 만들어 봅시다.

```python
file = open('Apple.txt', "r", encoding='UTF8')
text = file.read() 
file.close()
```

```python
########################## STEP1 - 데이터 전처리
wordList =text.split()  # 텍스트들을 단어 단위로 잘라 리스트에 담습니다.
 
worduniq = set(wordList) # Unique word 저장 : 중복된 값 제거
wordCount = {}           # 단어와 빈도수를 담을 빈 딕셔너리 공간.

for w in worduniq: 
    wordCount[w] = wordList.count(w) # 딕셔너리에 자른단어 갯수 저장 
# print(wordCount)

# 조사들은 제외합시다. 
del_word = ['in','had','you','You','was','to','not','I','it','we''in','In','the','a','is','of','on','that','this','and','be','will','into','our','have','from']
for w in del_word :
    if w in wordCount :
        del wordCount[w]
        
```

```python
wordCount
```

```python
# ②워드클라우드를 그립니다.
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# word cloud 그리기
wordcloud = WordCloud(
    width = 1000, height = 600).generate_from_frequencies(wordCount) # 우리가 만든딕셔너리
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.show()
```
